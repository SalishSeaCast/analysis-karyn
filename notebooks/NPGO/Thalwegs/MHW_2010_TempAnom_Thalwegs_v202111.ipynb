### Thalweg plot
  method using contour_thalweg from visualisations.py in tools repo

```python
#open bathy file and meshmask
fbathy=nc.Dataset('/ocean/ksuchy/MOAD/NEMO-forcing/grid/bathymetry_202108.nc')

fmesh=nc.Dataset('/ocean/ksuchy/MOAD/NEMO-forcing/grid/mesh_mask202108.nc')
```

```python
il=0
fig,ax=plt.subplots(1,1,figsize=(11,3))
#fig.title('Mesozooplankton Biomass')
levels = np.arange(-2, 2, .01)


cb=visualisations.contour_thalweg(ax,f1.variables['votemper'][il,...]-clim.variables['votemper'][0,...],fbathy,fmesh,clevels=levels,cmap=cmocean.cm.balance)
cb.set_label('$\mu$M', rotation=90,labelpad=6)
ax.set_ylim(430,0)

fig.suptitle('January 2010 Temperature Anomalies', fontsize=16)
#ax.text(15, 100, '(d)', fontsize=15,  color='k')
plt.savefig('TemperatureAnom201001.png', bbox_inches='tight',dpi=1000,transparent=False)
```
